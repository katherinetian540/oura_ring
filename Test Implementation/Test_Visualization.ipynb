```python
import pandas as pd
df = pd.read_csv('oura_2024-04-14_2024-04-28_trends.csv')
```

```python
import matplotlib.pyplot as plt

# Converting date to datetime
df['date'] = pd.to_datetime(df['date'])

# Sorting DataFrame by date
df = df.sort_values(by='date')

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(df['date'], df['Sleep Score'], marker='o', color='blue')
plt.title('Sleep Score')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Score')
 
plt.subplot(3, 2, 2)
plt.plot(df['date'], df['Average HRV'], marker='o', color='green')
plt.title('Average HRV Activity Score')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Score')

plt.subplot(3, 2, 3)
plt.plot(df['date'], df['Total Burn'], marker='o', color='red')
plt.title('Total Burn')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Burn')

plt.subplot(3, 2, 4)
plt.plot(df['date'], df['Resting Heart Rate Score'], marker='o', color='purple')
plt.title('Resting Heart Rate Score')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Score')

plt.subplot(3, 2, 5)
plt.plot(df['date'], df['Temperature Score'], marker='o', color='orange')
plt.title('Temperature Score')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Temperature (F)')

plt.subplot(3, 2, 6)
plt.plot(df['date'], df['Activity Score'], marker='o', color='green')
plt.title('Average HRV Activity Score')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Score')

plt.tight_layout()
plt.show()
```
